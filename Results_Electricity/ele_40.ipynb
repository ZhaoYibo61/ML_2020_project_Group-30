```
import numpy as np
import time

import pmdarima as pm

from BHT_ARIMA.util.utility import get_acc, nrmse

def run_ARIMA(data, param):


      order = param['order'] # order
      testsize = param['testsize']

      T = data.shape[-1]
      T_test = int((T * testsize) // 1)
      #T_test = min(T_test, 1)
      result_full = np.zeros([data.shape[0], T_test])

      total_time = 0
      n_round = 0


      for i in range(T_test):

        y = data[..., i:T-T_test+i].copy()
        n_round += 1
        start = time.time()

        for j in range(y.shape[0]):
          
          model = pm.AutoARIMA(seasonal=True, m=12, suppress_warnings=True,trace=True,error_action="ignore")#order

          result = model.fit_predict(y[j], n_periods=1)
          result_full[j, i] = result[..., -1]

        end = time.time()
        total_time = total_time + (end - start)

      true_value = data[..., -T_test:]


      stat = {}
      stat['acc'] = get_acc(result_full, true_value)
      stat['nrmse'] = nrmse(result_full, true_value)
      stat['ave_time'] = total_time/n_round


      return(stat)
```

```
import numpy as np

from statsmodels.tsa.api import VAR
from BHT_ARIMA.util.utility import get_acc, nrmse

def run_VAR(data):
  
      testsize = 0.1
      p = 1

      T = data.shape[-1]
      T_test = int((T * testsize) // 4)
      result_full = np.zeros([data.shape[0], T_test])

      total_time = 0
      n_round = 0

      for i in range(T_test):

        ts = (data[..., i:T-T_test+i].copy())
        n_round += 1
        model = VAR(ts)
        result = model.fit(p).forecast(ts, 1)
        result_full[..., i] = result[..., -1]

      label = data[..., -T_test:]

      stat = {}
      stat['acc'] = get_acc(result_full, label)
      stat['nrmse'] = nrmse(result_full, label)

      return(stat)
```

```
VAR_ele_40 = run_VAR(dataset_ele_40)
VAR_ele_40
```
